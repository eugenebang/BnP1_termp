```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('CNN_results_mrna.pkl', 'rb') as f :
    CNN_f, CNN_t,CNN_sc = pickle.load(f)
    
with open('HMM_results_mrna.pkl', 'rb') as f :
    HMM_f, HMM_t,HMM_sc = pickle.load(f)
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(9,7))
plt.plot(HMM_f, HMM_t, label='HMM : {:.3f}'.format(HMM_sc),c='r') #marker='.'
plt.plot(CNN_f, CNN_t, label='CNN : {:.3f}'.format(CNN_sc),c='c') #marker='.'
 
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(f'ROC curve of $Lin28a$ binding prediction models\n', size=17)
plt.legend()
plt.show()
```
